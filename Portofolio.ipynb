```python
database = pd.read_csv("D:\Kegiatan\Iramanesia\Streaming Data Dummy.csv")

database
```

```python
clean_data = database.dropna(inplace=False)

clean_data
```

```python
clean_data["Total Streams"] = (
    clean_data["Spotify Streams"]
    + clean_data["Youtube Music Plays"]
    + clean_data["Joox Streams"]
    + clean_data["Apple Music Streams"]
)

clean_data
```

```python
frameartismax = pd.DataFrame({
    "Most Stream Spotify": [clean_data.loc[clean_data["Spotify Streams"].idxmax(), "Nama"], f"{int(clean_data['Spotify Streams'].max()):,}"],
    "Most Stream Youtube": [clean_data.loc[clean_data["Youtube Music Plays"].idxmax(), "Nama"], f"{int(clean_data['Youtube Music Plays'].max()):,}"],
    "Most Stream Joox"   : [clean_data.loc[clean_data["Joox Streams"].idxmax(), "Nama"], f"{int(clean_data['Joox Streams'].max()):,}"],
    "Most Stream Apple"  : [clean_data.loc[clean_data["Apple Music Streams"].idxmax(), "Nama"], f"{int(clean_data['Apple Music Streams'].max()):,}"],
    "Most Stream Total"  : [clean_data.loc[clean_data["Total Streams"].idxmax(), "Nama"], f"{int(clean_data['Total Streams'].max()):,}"]
})

frameartismax.index = ["Artist Name", "Streams"]

frameartismax
```

```python
import matplotlib.pyplot as plt

# Ambil top 10
top10 = clean_data.sort_values(by="Total Streams", ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.bar(top10["Nama"], top10["Total Streams"], color="skyblue")
plt.xticks(rotation=45, ha="right")
plt.title("10 Most Played Artist Based On Total Streams")
plt.ylabel("Total Streams")
plt.show()
```

```python
frameartismax = pd.DataFrame({
    "Least Stream Spotify": [clean_data.loc[clean_data["Spotify Streams"].idxmin(), "Nama"], f"{int(clean_data['Spotify Streams'].min()):,}"],
    "Least Stream Youtube": [clean_data.loc[clean_data["Youtube Music Plays"].idxmin(), "Nama"], f"{int(clean_data['Youtube Music Plays'].min()):,}"],
    "Least Stream Joox"   : [clean_data.loc[clean_data["Joox Streams"].idxmin(), "Nama"], f"{int(clean_data['Joox Streams'].min()):,}"],
    "Least Stream Apple"  : [clean_data.loc[clean_data["Apple Music Streams"].idxmin(), "Nama"], f"{int(clean_data['Apple Music Streams'].min()):,}"],
    "Least Stream Total"  : [clean_data.loc[clean_data["Total Streams"].idxmin(), "Nama"], f"{int(clean_data['Total Streams'].min()):,}"]
})

frameartismax.index = ["Artist Name", "Total Streams"]

frameartismax
```
